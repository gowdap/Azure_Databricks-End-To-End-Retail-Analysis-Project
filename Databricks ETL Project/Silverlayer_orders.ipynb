```python
from pyspark.sql.functions import *
```

```python
df = spark.read.format("parquet").load("abfss://bronze@databricksstorageadls.dfs.core.windows.net/orders")
df.display()
```

```python
df.printSchema()
```

```python
df = df.drop("_rescued_data")

```

```python
df = df.withColumn("order_date", to_timestamp(col("order_date")))
df.display()
```

```python
df = df.withColumn("year", year(col("order_date")))
df.display()
```

```python
df.write.format("delta").mode("overwrite").save("abfss://silver@databricksstorageadls.dfs.core.windows.net/orders")
```

```python
%sql
create table if not exists databricks_cata.silver.orders_silver
using delta
location "abfss://silver@databricksstorageadls.dfs.core.windows.net/orders"
```
