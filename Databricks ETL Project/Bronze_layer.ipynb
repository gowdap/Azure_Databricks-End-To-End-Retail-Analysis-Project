### Dynamic capabilities

```python
dbutils.widgets.text("file_name", "")
```

```python
p_file_name= dbutils.widgets.get("file_name")
```

```python
df = spark.read.format("parquet").load(f"abfss://bronze@databricksstorageadls.dfs.core.windows.net/{p_file_name}")
df.display()
```

### Data Reading


```python
df = spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format", "parquet")\
        .option("cloudFiles.schemaLocation", f"abfss://bronze@databricksstorageadls.dfs.core.windows.net/checkpoint_{p_file_name}")\
            .load(f"abfss://source@databricksstorageadls.dfs.core.windows.net/{p_file_name}")
```

### Data writing

```python
df.writeStream.format("parquet")\
    .outputMode("append")\
        .option("checkpointLocation", f"abfss://bronze@databricksstorageadls.dfs.core.windows.net/checkpoint_{p_file_name}")\
            .option("path", f"abfss://bronze@databricksstorageadls.dfs.core.windows.net/{p_file_name}")\
                .trigger(once=True)\
                    .start()
```
